```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/Music notes Dataset.csv')
data.sample(5)
```

```python
data.shape
```

```python
input = data.drop('label', axis = 1)
output = data['label']
```

```python
# plt.figure(figsize = (7, 7))
```

```python
index = 434
grid_data = input.iloc[index].values.reshape(64, 64)
plt.imshow(grid_data, interpolation = None, cmap = "Blues")
print(output[index])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input, output, test_size = 0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
model.predict(x_test[:5])
```

```python
data["Predicted Label"] = model.predict(input)
```
